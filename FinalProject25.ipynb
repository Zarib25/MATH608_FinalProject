```python
study_exam = studentPF[["Hours_Studied", "Exam_Score", "Attendance", "Tutoring_Sessions"]]
study_exam
```

```python
#scatterplot to show if there is a relationship between hours studied and exam scores
study_exam_plot = alt.Chart(study_exam).mark_circle().encode(
    x=alt.X("Hours_Studied").title("Hours studied per week"),
    y=alt.Y("Exam_Score").title("Exam Score").scale(zero=False),
    tooltip=["Hours_Studied", "Exam_Score"]
).properties(
    title="Relationship between hours studied weekly and Exam Scores.").interactive()
study_exam_plot
```

```python
# Visualizing Attendance within the Study/Exam relationship
outlier_plot = alt.Chart(study_exam).mark_circle().encode(
    x=alt.X("Hours_Studied:Q", title="Weekly Study Hours").scale(zero=False),
    y=alt.Y("Exam_Score:Q", title="Exam Score").scale(zero=False),
    
# Adding Attendance as a color gradient to see its impact
    color=alt.Color("Attendance:Q", scale=alt.Scale(scheme='viridis'), title="% of Attendance"),
    
# Adding tooltips 
    tooltip=["Attendance", "Hours_Studied", "Exam_Score"]
).properties(
    title="Impact of Attendance on High-performing, low study group of student"
).interactive()

outlier_plot
```

```python
# Base scatter plot
reg_line = alt.Chart(study_exam).mark_circle().encode(
    x="Hours_Studied:Q",
    y="Exam_Score:Q",
    tooltip=["Hours_Studied", "Exam_Score"]
)

# Layered trend line (shows slope and strength)
line = reg_line.transform_regression("Hours_Studied", "Exam_Score").mark_line(color='red')

(reg_line + line).properties(title="Analyzing Slope and Strength of relationship").interactive()
```

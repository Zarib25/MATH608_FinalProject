```python
import pandas as pd
from scipy.io import arff

data = arff.loadarff("Data/Autism-Child-Data.arff")
autism_data = pd.DataFrame(data[0])
autism_data
```

```python
autism_data.info()
```

```python
autCols = [col for col in autism_data.columns if autism_data[col].dtype=="O"]
autCols[:5]
```

```python
autism_data[autCols]=autism_data[autCols].apply(lambda x: x.str.decode('utf8'))
autism_data.head()
```

```python
autism_data.to_csv('data/autism_data.csv',index=False)
```

```python
import pandas as pd
from scipy.io import arff

def arff_to_csv_pandas("", csv_file_path):
    """
    Converts an ARFF file to a CSV file using pandas and scipy.io.arff.

    Args:
        arff_file_path (str): The path to the input ARFF file.
        csv_file_path (str): The path to the output CSV file.
    """
    try:
        data = arff.loadarff(arff_file_path)
        df = pd.DataFrame(data[0])
        df.to_csv(csv_file_path, index=False)
        print(f"Successfully converted '{arff_file_path}' to '{csv_file_path}'")
    except Exception as e:
        print(f"Error converting ARFF to CSV: {e}")

# Example usage:
# arff_to_csv_pandas('your_file.arff', 'output_file.csv')
```
